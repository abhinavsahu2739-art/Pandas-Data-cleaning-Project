```python
## Q11 Missing values in each column
df.isnull().sum()
```

```python
## Q12 customers with missing purchase _amount
df["Purchase_Amount"].isnull().sum()
```

```python
##  Q13 Replace missing purchase amount
mean_purchased_amount=df["Purchase_Amount"].mean()
df.fillna(mean_purchased_amount,inplace=True)
df["Purchase_Amount"].isnull().sum()
```

```python
#Q14 missing values in city column
df["City"].isnull().sum()
```

```python
#Q 15 remove row where age is missing
df.dropna(subset=["Age"],inplace=True)
df
```

```python
# Q16 standardize gender column
# Convert Gender column to uppercase for consistent processing
df["Gender"] = df["Gender"].astype(str).str.upper()

# Replace variations with 'MALE' or 'FEMALE'
df["Gender"] = df["Gender"].replace({
    "M": "MALE",
    "F": "FEMALE",
    "NAN": "UNKNOWN" # Replace string "NAN" if it exists
})
```

```python
df
```

```python
#17 clean city column
df["City"].str.strip().str.upper()
```

```python
#Q18 standardize country column to india
df["Country"]=df["Country"].str.strip()
df["Country"].unique()
df.loc[df["Country"].str.upper()=="INDIA"]
```

```python
#Q19 convert signupdate to datetime
df["Signup_Date"] = pd.to_datetime(df["Signup_Date"])
df
```

```python
#Q20 convert lastpurchase date to datetime
df["Last_Purchase_Date"]=pd.to_datetime(df["Last_Purchase_Date"])
df
```

```python
#Q21 average purchase amount
df["Purchase_Amount"].mean()
```

```python
# q22 city wise average purchase amount
df.groupby("City")["Purchase_Amount"].mean()
```

```python
#Q23customers count per city
df["City"].value_counts()
```

```python
#Q24 customers with purchase amount above overall average
df[df["Purchase_Amount"]>df["Purchase_Amount"].mean()]
```

```python
#q25 create customer type column
def customer_type(x):
  if x>30000 :
    return "High value customer"
  else:
    return "Regular"
```

```python
# apply
df["new_customer"]=df["Purchase_Amount"].apply(customer_type)
```
